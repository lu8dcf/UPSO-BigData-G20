```python
import os
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# BUSQUEN LA RUTA DONDE ESTAN LOS CSV - EN DROPBOX BIG DATA

# ruta lucia
#ruta = 'C:/Users/luxam/nubes/Dropbox/Estudio/Upso/GRUPO-TRABAJO/BigData/Laboratorio1/Datos/'

# ruta heber
ruta= 'D:/Dropbox/Estudio/Upso/GRUPO-TRABAJO/BigData/Laboratorio1/Datos/'

# ruta viviana
#ruta= '/content/drive/MyDrive/BIG_DATA/DATASET_MESES/Datos'

# ruta pastor
#ruta='C:/Users/pasto/Dropbox/GRUPO-TRABAJO/BigData/Laboratorio1/Datos'


archivos = os.listdir(ruta)
archivos


dataframes ={}
# df_completo = pd.DataFrame()

for archivo in archivos:
    if archivo.endswith('.csv'):  #Asegurarse de que se consideren archivos csv
      mes = archivo.split('_')[3].split('.')[0]  # Obtener el nombre del mes sin la extensión
      ruta_completa = os.path.join(ruta,archivo) # Ruta de acceso completa
      df = pd.read_csv(ruta_completa) # Cargar el csv en un Dataframe
      df_ventas = df.copy()  # Hago copia

      # Transformo todos los valores no numericos en Nan en las columnas Cantidad Pedida y Precio Unitario
      df_ventas['Cantidad Pedida'] = pd.to_numeric(df_ventas['Cantidad Pedida'], errors='coerce')  #El errors= 'coerce' transforma los valores no numericos a NaN'
      df_ventas['Precio Unitario'] = pd.to_numeric(df_ventas['Precio Unitario'], errors='coerce')

      # Elimino todos los Nan y me quedo solo con los valores numericos
      df_ventas = df_ventas.loc[~df_ventas['Cantidad Pedida'].isna()]
      df_ventas = df_ventas.loc[~df_ventas['Precio Unitario'].isna()]

      df_ventas = df_ventas.dropna()

      # Transformo todos los valores numericos al tipo de dato int
      df_ventas['Cantidad Pedida'] = df_ventas['Cantidad Pedida'].astype(int)
      df_ventas['Precio Unitario'] = df_ventas['Precio Unitario'].astype(float)


      dataframes[mes] = df_ventas

df_ventas
```

```python
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
dias = list(range(1, 32))

totales_ventas =[] # Ventas Por mes
cantidad_ventas =[] # Cantidad de productos vendidos por mes

for mes in meses:

  dataframe_mes = dataframes[mes]
  dataframe_mes['Total venta'] = dataframe_mes['Cantidad Pedida'] * dataframe_mes['Precio Unitario'] # Genera una columna con la venta total del item
  
  cantidad_ventas_mes = int(dataframe_mes['Cantidad Pedida'].sum())
  cantidad_ventas.append(cantidad_ventas_mes)

  totales_ventas_mes = int(dataframe_mes['Total venta'].sum())  # suma todas las ventas del mes
  totales_ventas.append(totales_ventas_mes) # agrega a la lista la venta total del mes correspondiente
  
  
cantidad_ventas, totales_ventas
```

```python
# funcion para separa fecha y hora de los pedidos para poder analizarlos
def separ_datos(datos):
    datos['Fecha de Pedido']=pd.to_datetime(dataframe_mes['Fecha de Pedido']) #transformo datos de fecha en un formato legible por pandas
    datos['Hora']=datos['Fecha de Pedido'].dt.hour
    datos['Dia']=datos['Fecha de Pedido'].dt.day
    return datos
```

```python
Horas = list(range(24)) # defino rango horario

fig = go.Figure()

for mes in meses:
  dataframe_mes = dataframes[mes]

  separ_datos(dataframe_mes) #llamo a la funcion para separa dia y hora
  grupos=dataframe_mes.groupby('Hora')['Cantidad Pedida'].sum() #agrupo por hora y sumo la cantidad pedida por hora

  hora_max_venta = grupos.idxmax()  # Obtiene la hora de mayor venta en el mes actual

# grafico, por cada iteracion del ciclo for (cada mes), me genera una linea en mi grafico identificada con distintos colores por cada mes
  fig.add_trace(go.Scatter(
    x = Horas,
    y = grupos,
    mode = 'lines',
    name = mes,
    line = dict(shape='spline', smoothing=1.3, width=2),
    text=f'Hora máxima de venta: {hora_max_venta}',  # Agrega el texto con la hora máxima de venta en el mes actual
    ))

fig.update_layout(
    title= "Unidades vendidas",
    xaxis=dict(title="Hora", tickvals= Horas),
    yaxis=dict(title="unidades vendidas", side="left"))
#imprime un solo grafico con cada uno de los meses del ciclo for anterior
fig.show()
```
